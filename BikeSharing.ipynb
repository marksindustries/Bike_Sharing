# Bike Sharing 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
train = pd.read_csv("Desktop/bike_sharing_train.csv")
```

```python
train.head()
```

```python
train.info()
```

```python
train["datetime"] = pd.to_numeric(pd.to_datetime(train["datetime"]))
```

```python
train.head()
```

### 1) Avg count based on the season.

```python
train.groupby(["season"])["count"].mean().plot(kind="bar",figsize=(8,5),title="Avg count based on Seasons")
```

```python
train.groupby(["holiday"])["count"].mean().plot(kind="bar")
```

```python
train.groupby(["workingday"])["count"].mean().plot(kind="bar")
```

```python
train.groupby(["weather"])["count"].mean().plot(kind="bar")
```

```python
train.groupby(["weather"])["temp"].agg({"mean","max","min"}).plot(kind="bar",figsize=(10,8))
```

```python
train.groupby(["season","holiday"])["count"].mean().unstack().plot(kind="bar",figsize=(10,8))
```

```python
train.groupby(["season","weather"])["temp"].mean().unstack().plot(kind="bar",figsize=(10,8))
```

```python
train.groupby(["season","weather"])[["temp"]].mean().unstack().plot(kind="bar",figsize=(10,8))
```
